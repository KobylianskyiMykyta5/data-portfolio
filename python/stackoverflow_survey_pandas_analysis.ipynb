```python
import pandas as pd
```

```python
df_public = pd.read_csv('survey_results_public.csv')
df_schema = pd.read_csv('survey_results_schema.csv')
```

```python
df_public.sample(5)
```

```python
df_schema.sample(5)
```

```python
# 1. Скільки респондентів пройшли опитування?
print('Count of respondents: ', df_public.shape[0])
```

```python
# 2. Скільки респондентів відповіли на всі запитання?
questions_set = set(df_schema['qname']) & set(df_public.columns)
print('Answered on all questions: ', df_public[list(questions_set)].dropna().shape[0])
```

```python
# 3. Які значення мір центральної тенденції для досвіду (WorkExp) респондентів?

df_public['WorkExp'] = pd.to_numeric(df_public['WorkExp'], errors='coerce')

print('AVG_exp: ', round(df_public['WorkExp'].mean(),1))
print('Median_exp: ', round(df_public['WorkExp'].median(),1))
print('Mode_exp: ', round(df_public['WorkExp'].mode(),1))
```

```python
# 4. Скільки респондентів працює віддалено?
print('Respondents work remotely: ', (df_public['RemoteWork'] == 'Remote').sum())
```

```python
# 5. Який відсоток респондентів програмує на Python?
python_users = df_public['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()
percent = (python_users / df_public.shape[0]) * 100
print(f'Respondents who have worked with Python: {percent:.2f}%')
```

```python
# 6. Скільки респондентів навчалося програмувати за допомогою онлайн курсів?
learned_online = (df_public['LearnCodeOnline'] == 'Online courses').sum()
print(f'Learn code online: {learned_online}')
```

```python
# 7. Серед респондентів що програмують на Python в групуванні по країнам, яка середня та медіанна сума компенсації (ConvertedCompYearly) в кожній країні?
df_public['is_python_user'] = df_public['LanguageHaveWorkedWith'].str.contains('Python', na=False)
avg_med_ccy =  round(df_public[df_public['is_python_user']].groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']), 1)
avg_med_ccy
```

```python
# 8. Які рівні освіти мають 5 респондентів з найбільшою компенсацією?
sorted_df = df_public.sort_values(by='ConvertedCompYearly', ascending=False)
top5_ed = sorted_df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']].head(5)
top5_ed
```

Бонусні запитання:

```python
# 1. В кожній віковій категорії, який відсоток респондентів програмує на Python?
df_public['is_python_user'] = df_public['LanguageHaveWorkedWith'].str.contains('Python', na=False)
percent_by_age = (df_public.groupby('Age')['is_python_user'].mean() * 100)
percent_by_age
```

```python
# 2. Серед респондентів що знаходяться в 75 перцентилі за компенсацією середнього і працюють віддалено, які індустрії є найрозповсюдженішими?
q75 = df_public['ConvertedCompYearly'].quantile(0.75)
remote_top = df_public[(df_public['RemoteWork'] == 'Remote') & (df_public['ConvertedCompYearly'] >= q75)]
top5_industry = remote_top['Industry'].dropna().value_counts()
top5_industry
```
